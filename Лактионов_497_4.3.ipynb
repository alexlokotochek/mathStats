```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
N = 1000
bootstraps = np.arange(500)
xLine = np.arange(0, 1, .01)
```

```python
variance = np.zeros(len(xLine)) 

for i in xLine:
    xFirst = np.random.binomial(1, i, N)
    theta = np.mean(xFirst)
    a = b = 0
    for j in bootstraps:
        x = np.random.binomial(1, theta, N)
        y = np.mean(x)
        a += y
        b += y**2.   
    index = (int)(i*100 + 1e-8)
    variance[index] = (b/500.-(a/500.)**2.)

plt.plot(xLine, variance)
f = lambda x: (x * (1 - x) / N)
plt.plot(xLine, f(xLine))
```
