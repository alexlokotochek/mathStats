```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from mpl_toolkits.mplot3d.axes3d import Axes3D
import scipy.stats as st
%matplotlib inline
```

```python
alpha = 0.95
N = 100
```

```python
X = numpy.random.normal(0, 1, N)
```

Построим выборку $X_1, X_2, ... , X_n$ из $N(0, 1)$.

```python
def build_graph_interval(min_theta, max_theta, a, b):
    x0 = numpy.arange(1, N + 1, 1)
    y1 = numpy.zeros(N)
    y2 = numpy.zeros(N)

    for n in range(1, N + 1):
        y1[n - 1] = min_theta(X[:n], n)
        y2[n - 1] = max_theta(X[:n], n)

    fig = plt.figure(figsize=(20, 10))
    ax = fig.gca()
    plt.fill_between(x0, y1, y2)

    ax.set_xlim(0, N)
    ax.set_ylim(a, b)

    plt.show()
```

```python
build_graph_interval((lambda x, n: np.mean(x) - st.norm.ppf(0.5 + alpha / 2) * n ** -0.5), 
                     (lambda x, n: np.mean(x) + st.norm.ppf(0.5 + alpha / 2) * n ** -0.5),
                     -1, 1) # , loc=0, scale=n**-0.5)
```

Найдем точный доверительный интервал для $a$ при $\sigma^2 = 1$.

$\overline X  - a$ распределен как $N(0, \frac{1}{n})$.

$P(-z_{\frac{1+\alpha}{2}} < (\overline X  - a) * \sqrt n< z_{\frac{1+\alpha}{2}}) = \alpha$, где $z_{\frac{1+\alpha}{2}$ - квантиль стандратного нормального распределения.

$P(\overline X - \frac{z_{\frac{1+\alpha}{2}}}{\sqrt n} < a < \overline X + \frac{z_{\frac{1+\alpha}{2}}}{\sqrt n}) = \alpha$

```python
build_graph_interval((lambda x, n: numpy.sum(x * x) / st.chi2.ppf(0.5 + alpha / 2, df=n)), 
                     (lambda x, n: numpy.sum(x * x) / st.chi2.ppf(0.5 - alpha / 2, df=n)),
                     0, 2)
```

Найдем точный доверительный интервал для $\sigma^2$ при $a = 0$.

$\overline {X^2} $ распределен как $\chi^2(n) \sigma^2$ 

$P(z_{\frac{1-\alpha}{2}} < \frac{\overline {X^2}}{\sigma^2} < z_{\frac{1+\alpha}{2}}) = \alpha$, где $z$ - квантиль распределения $\chi^2(n)$

$P(\frac{\overline {X^2}}{z_{\frac{1-\alpha}{2}}} < \sigma^2 < \frac{\overline {X^2}}{z_{\frac{1+\alpha}{2}}}) = \alpha$


```python
build_graph_interval((lambda x, n: np.mean(x) - st.t.ppf(0.5 + alpha / 2, n - 1) * np.sum([abs(xi - np.mean(x)) for xi in x]) *
                                   (n * ((n - 1) if n > 1 else 0.0001)) ** -0.5), 
                     (lambda x, n: np.mean(x) + st.t.ppf(0.5 + alpha / 2, n - 1) * np.sum([abs(xi - np.mean(x)) for xi in x]) *
                                   (n * ((n - 1) if n > 1 else 0.0001)) ** -0.5),
                     -3, 3)
```

Найдем точный доверительный интервал для $a$ при неизвестном $\sigma^2$.

Используем факт из лекций о точном интервале для метода наименьших квадратов:

$P(-z_{\frac{1+\alpha}{2}} < \frac{(\overline X - a) \sqrt{n(n-1)}}{|X - \overline X|} < z_{\frac{1+\alpha}{2}}) = \alpha$, где $z$ - квантиль распределения $T(n - 1)$

$P(\overline X - \frac{z_{\frac{1+\alpha}{2}} |X - \overline X|}{\sqrt{n(n-1)}}< a < \overline X + \frac{z_{\frac{1+\alpha}{2}} |X - \overline X|}{\sqrt{n(n-1)}}) = \alpha$


```python
build_graph_interval((lambda x, n: 0), 
                     (lambda x, n: n * np.var(x) / st.chi2.ppf(1 - alpha, df=n-1)),
                     0, 3)
```
