```python
import math
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
from sympy import *
from matplotlib import cm
from scipy.stats import multivariate_normal
from matplotlib import rcParams
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from scipy import integrate
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
gamma = 0.95
N = 100
```

```python
def build_graph(f1,f2,gamma,X,title,low,high,left=1):
    A_X = np.arange(left,N+1)
    F1 = [f1(X[0:i+1],gamma) for i in range(len(X))]
    F2 = [f2(X[0:i+1],gamma) for i in range(len(X))]
    
    plt.figure(figsize=(15,8))
    plt.grid(True)
    plt.title(title,fontsize=15)
    plt.plot(A_X,F1,color='r')
    plt.plot(A_X,F2,color='b')
    plt.fill_between(A_X,F1,F2,facecolor='purple')
    plt.axis([left,len(X),low,high])
    plt.xlabel(r'$n$')
    plt.ylabel(r'$\theta$')
    plt.show()
```

```python
def get_interval_prob(numb,f1,f2,gen,theta,check):
    correct =[0 for i in range(len(check))]
    for i in range(numb):
        X = gen()
        for j in range(len(check)):
            if  f1(X[0:check[j]],gamma) < theta < f2(X[0:check[j]],gamma):
                correct[j]+=1
    return [round(it/numb,3) for it in correct]
```

Для построения точных доверительных интервалов будем использовать следующие утверждения:
- $\frac{\bar{X} - a}{\sigma}\sqrt{N} \sim \mathcal{N}(0,1)$
- $\frac{\bar{X} - a}{\tilde{s}}\sqrt{N} \sim t_{n-1}$,где  $t_{n-1}$ распределение Стьюдента с $n-1$ степенью свободы,
$\tilde{s} = \sqrt{\frac{\sum_{i=1}^{n}(X_i - \bar{X})^2}{n-1}}$ - стандартное отклонение
- $\frac{(n-1)\tilde{s}^2}{\sigma^2} \sim \chi^2_{n-1}$
- $\frac{ns^2}{\sigma^2} \sim \chi^2_{n}$, где $s = \sqrt{\frac{\sum_{i=1}^{n}(X_i - \bar{X})^2}{n}}$ 

Для $\mathcal{N}(\theta,1)$ точным доверительным интервалом уровня $\gamma$ будет:

- $(\bar{X} - Z_{\frac{1+\gamma}{2}}\frac{1}{\sqrt{N}}, \bar{X} + Z_{\frac{1+\gamma}{2}}\frac{1}{\sqrt{N}})$

где $Z_{(\frac{1+\gamma}{2})}$ - квантиль нормального распределения $\mathcal{N}(0,1)$ уровня $\frac{1 + \gamma}{2}$

```python
X = sps.norm.rvs(size=N)
Z = sps.norm.ppf((1 + gamma)/2)
build_graph((lambda X,gamma: np.mean(X) - Z * len(X)**(-0.5)),
            (lambda X,gamma: np.mean(X) + Z * len(X)**(-0.5)), gamma, X, 
            'trust interval for ' + r'$\mathcal{N}(\theta, 1), \theta = 0$', -1,1)
```

```python
# проверка
get_interval_prob(10000, (lambda X,gamma: np.mean(X) - Z * (1/(len(X)))**0.5),
            (lambda X,gamma: np.mean(X) + Z * (1/(len(X)))**0.5),
                 (lambda : sps.norm.rvs(size=N)), 0,[10,100])
```

Для $\mathcal{N}(0,\theta)$ точным доверительным интервалом уровня $\gamma$ будет:

- ($\frac{\sum_{i=1}^{n}{X_i^2}}{\nu_{\frac{1+\alpha}{2}}},\frac{\sum_{i=1}^{n}{X_i^2}}{\nu_{\frac{1-\alpha}{2}}})$, где $\nu$ квантиль распределения $\chi^2_n$

```python
X = sps.norm.rvs(size=N)
build_graph((lambda X,gamma: np.sum(X**2)/sps.chi2.ppf(df=len(X),q=(1 + gamma)/2)),
            (lambda X,gamma: np.sum(X**2)/sps.chi2.ppf(df=len(X),q=(1 - gamma)/2)), gamma, X, 
            'trust interval for ' + r'$\mathcal{N}(0, \theta), \theta = 1$', -1,4)
```

```python
# проверка
get_interval_prob(10000,(lambda X,gamma: np.sum(X**2)/sps.chi2.ppf(df=len(X),q=(1 + gamma)/2)),
            (lambda X,gamma: np.sum(X**2)/sps.chi2.ppf(df=len(X),q=(1 - gamma)/2)) ,
                 (lambda : sps.norm.rvs(size=N)), 1,[10,100])
```

Для $\mathcal{N}(\theta,\sigma^2)$ точным доверительным интервалом уровня $\gamma$ будет:

- $(\bar{X} - t_{\frac{1+\gamma}{2}}\frac{\tilde{s}}{\sqrt{N}}, \bar{X} + t_{\frac{1+\gamma}{2}}\frac{\tilde{s}}{\sqrt{N}})$

где $t_{(\frac{1+\gamma}{2})}$ - квантиль распределения Стьюдента с $n-1$ степенью свободы уровня $\frac{1 + \gamma}{2}$,
$\tilde{s} = \sqrt{\frac{\sum_{i=1}^{n}(X_i - \bar{X})^2}{n-1}}$ - стандартное отклонение

```python
X = sps.norm.rvs(size=N)
build_graph((lambda X, gamma: np.mean(X) - (np.var(X) / (len(X)-1)) ** 0.5 * sps.t.ppf((1 + gamma) / 2, df=len(X)-1)),
            (lambda X, gamma: np.mean(X) + (np.var(X) / (len(X)-1)) ** 0.5 * sps.t.ppf((1 + gamma) / 2, df=len(X)-1)), 
            gamma, X, 'trust interval for ' + r'$\mathcal{N}(\theta, \sigma^2), \theta = 0$', -2,2)
```

```python
# проверка
get_interval_prob(10000,
            (lambda X, gamma: np.mean(X) - (np.var(X) / (len(X)-1)) ** 0.5 * sps.t.ppf((1 + gamma) / 2, df=len(X)-1)),
            (lambda X, gamma: np.mean(X) + (np.var(X) / (len(X)-1)) ** 0.5 * sps.t.ppf((1 + gamma) / 2, df=len(X)-1)) ,
            (lambda : sps.norm.rvs(size=N)), 0,[10,100])
```

Для $\mathcal{N}(a,\theta)$ точным доверительным интервалом уровня $\gamma$ будет:

- ($\frac{(n-1)\tilde{s}}{\mu_{\frac{1+\alpha}{2}}}, \frac{(n-1)\tilde{s}}{\mu_{\frac{1-\alpha}{2}}})$, где $\nu$ квантиль распределения $\chi^2_{n-1}$, $\tilde{s} = \sqrt{\frac{\sum_{i=1}^{n}(X_i - \bar{X})^2}{n-1}}$ - стандартное отклонение

```python
X = sps.norm.rvs(size=N)
build_graph((lambda X,gamma: len(X)*np.var(X)/sps.chi2.ppf(df=len(X)-1,q=(1 + gamma)/2)),
            (lambda X,gamma: len(X)*np.var(X)/sps.chi2.ppf(df=len(X)-1,q=(1 - gamma)/2)), gamma, X, 
            'trust interval for ' + r'$\mathcal{N}(0, \theta), \theta = 1$', -1,4)
```

```python
# проверка
get_interval_prob(10000,
            (lambda X,gamma: len(X)*np.var(X)/sps.chi2.ppf(df=len(X)-1,q=(1 + gamma)/2)),
            (lambda X,gamma: len(X)*np.var(X)/sps.chi2.ppf(df=len(X)-1,q=(1 - gamma)/2)),
            (lambda : sps.norm.rvs(size=N)), 1,[10,100])
```
